```python
# function that returns only the names of the neighbourhoods without the numbers
# column "neigb_cijf" is used
def get_only_names(df, column_name): #column_name must not be a string
    names = []
    for name in df[column_name]:
        names_aux = name.split(' ') #splits each entrace of "neigb_cijf" column based on the spaces between elements
        names_aux.pop(0) #removes the first element of the newly created list wich is a the number to remove
        names_aux_joined = ' '.join(names_aux) #joins the other elements on the list (this list no longer contains the number)
        names.append(names_aux_joined)
    del df[column_name]
    df["Neighbourhood"] = names
```
